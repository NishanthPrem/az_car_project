```python
from delta.tables import DeltaTable
```

# Reading Silver Data

```python
df_silver = spark.sql("SELECT * FROM PARQUET.`abfss://silver@nishcarsdatalake.dfs.core.windows.net/car_sales`")
```

```python
df_model = spark.sql("SELECT * FROM cars_catalog.gold.dim_model")
df_branch = spark.sql("SELECT * FROM cars_catalog.gold.dim_branch")
df_dealer = spark.sql("SELECT * FROM cars_catalog.gold.dim_dealer")
df_date = spark.sql("SELECT * FROM cars_catalog.gold.dim_date")
```

```python
df_fact = df_silver.join(df_model, df_silver.Model_ID == df_model.Model_ID, "left")\
                    .join(df_branch, df_silver.Branch_ID == df_branch.Branch_ID, "left")\
                    .join(df_dealer, df_silver.Dealer_ID == df_dealer.Dealer_ID, "left")\
                    .join(df_date, df_silver.Date_ID == df_date.Date_ID, "left")\
                    .select(df_silver["Revenue"], df_silver["Units_Sold"], df_silver["Revenue_per_unit"], df_model["dim_key"], df_branch["dim_branch_key"], df_dealer["dim_dealer_key"], df_date["dim_date_key"])
```

```python
display(df_fact)
```

```python
if spark.catalog.tableExists("cars_catalog.gold.fact_sales"):
    delta_table = DeltaTable.forName(spark, "cars_catalog.gold.fact_sales")
    delta_table.alias("trg").merge(df_fact.alias("src"), "trg.dim_key = src.dim_key and trg.dim_branch_key = src.dim_branch_key and trg.dim_dealer_key = src.dim_dealer_key and trg.dim_date_key = src.dim_date_key").whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
else:
    df_fact.write.mode("overwrite").option("path", "abfss://gold@nishcarsdatalake.dfs.core.windows.net/fact_sales").saveAsTable("cars_catalog.gold.fact_sales")
```
